# Magenta RealTime 2 — `flax.nnx` backend demo

Generate a few seconds of audio with the `flax.nnx` inference backend
(`MagentaRT2Nnx`), which exposes the same system API as the `jax` / `mlx`
backends: `embed_style` → `generate` → `(AudioTree, state)`.

Requires `pip install magenta-rt[jax]` and a downloaded checkpoint
(`mrt models init`, then `mrt checkpoints download`).

```python
from magenta_rt import MagentaRT2Nnx

# Builds the nnx system and loads mrt2_small.safetensors from the checkpoints dir.
mrt = MagentaRT2Nnx(size="mrt2_small")
```

```python
# Embed a text prompt into a MusicCoCa style embedding, then stream audio.
# Each frame is 40 ms, so 25 frames is ~1 second.
style = mrt.embed_style("disco funk")
audio_tree, state = mrt.generate(style=style, frames=25)
print("waveform:", audio_tree.waveform.shape, "sample_rate:", audio_tree.sample_rate)

audio_tree.write("demo_magenta_rt_v2_nnx.wav")
```

```python
import IPython.display as ipd
import numpy as np

# AudioTree.waveform is [batch, channels, samples]; play the first clip.
clip = np.asarray(audio_tree.waveform[0])  # [channels, samples]
ipd.Audio(clip, rate=audio_tree.sample_rate)
```
